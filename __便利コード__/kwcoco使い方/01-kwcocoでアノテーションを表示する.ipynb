# COCOデータセットのアノテーションを描画して表示

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import kwcoco, kwplot
import random

coco = kwcoco.CocoDataset("../../sample_data/coco_sample_datasets/sample_coco_train2017/instances_train2017.json")

sample_imgId = random.choice(list(coco.imgs.keys()))
print(sample_imgId)

kwplot.autompl()
coco.show_image(gid=sample_imgId)
```

# アノテーションを描画した画像データを取得して表示

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import kwcoco, kwplot
import random
from matplotlib import pyplot as plt
import cv2

coco = kwcoco.CocoDataset("../../sample_data/coco_sample_datasets/sample_coco_train2017/instances_train2017.json")

sample_imgId = random.choice(list(coco.imgs.keys()))
print(sample_imgId)

kwplot.autompl()
visualized_image = coco.draw_image(gid=sample_imgId)

kwplot.autompl()
kwplot.imshow(visualized_image)
```
