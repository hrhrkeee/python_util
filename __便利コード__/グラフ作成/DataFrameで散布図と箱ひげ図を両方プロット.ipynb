```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.gridspec as gridspec
import random


def plot_data(df, ax_label_X, ax_label_Y, figsize=(12, 10), xlim=(-0.05, 1.05), ylim=(-0.05, 1.05)):
    
    def annotate_stats(ax, data, orient, fontsize=10):

        if orient == 'vertical':
            quartiles = data.quantile([0.25, 0.5, 0.75]).values
            vertical_spacing = (data.max() - data.min()) / 4  # Adjust spacing based on data range
            vertical_spacing = 0  # Adjust spacing based on data range
            ax.text(1, data.max()   + 3*vertical_spacing,     f'Max: {data.max():.2f}',                   fontsize=fontsize)
            ax.text(1, quartiles[2] + 2*vertical_spacing,     f'Q3: {quartiles[2]:.2f}',                  fontsize=fontsize)
            ax.text(1, data.mean()  + 1*vertical_spacing,   f'Mean: {data.mean():.2f}',                 fontsize=fontsize)
            ax.text(1, data.mean()  - 1*vertical_spacing,   f'Median: {quartiles[1]:.2f}',                 fontsize=fontsize)
            ax.text(1, quartiles[0] - 2*vertical_spacing,     f'Q1: {quartiles[0]:.2f}',                  fontsize=fontsize)
            ax.text(1, data.min()   - 3*vertical_spacing,     f'Min: {data.min():.2f}',                   fontsize=fontsize)
        else:
            quartiles = data.quantile([0.25, 0.5, 0.75]).values
            horizontal_spacing = (data.max() - data.min()) / 4  # Adjust spacing based on data range
            horizontal_spacing = 0  # Adjust spacing based on data range
            ax.text(data.max()   + 3*horizontal_spacing, 1.8,   f'Max: {data.max():.2f}',      rotation=90,      fontsize=fontsize)
            ax.text(quartiles[2] + 2*horizontal_spacing, 1.8,   f'Q3: {quartiles[2]:.2f}',     rotation=90,     fontsize=fontsize)
            ax.text(data.mean()  + 1*horizontal_spacing, 1.8, f'Mean: {data.mean():.2f}',    rotation=90,    fontsize=fontsize)
            ax.text(data.mean()  - 1*horizontal_spacing, 1.8, f'Median: {quartiles[1]:.2f}', rotation=90)
            ax.text(quartiles[0] - 2*horizontal_spacing, 1.8,   f'Q1: {quartiles[0]:.2f}',     rotation=90,     fontsize=fontsize)
            ax.text(data.min()   - 3*horizontal_spacing, 1.8,   f'Min: {data.min():.2f}',      rotation=90,      fontsize=fontsize)

    
    # グリッドレイアウトの設定
    fig = plt.figure(figsize=figsize)
    gs = gridspec.GridSpec(3, 3, height_ratios=[1, 8, 1], width_ratios=[1, 8, 1])

    # 散布図
    ax = fig.add_subplot(gs[1, 1])
    sns.scatterplot(data=df, x=ax_label_X, y=ax_label_Y, ax=ax)
    ax.grid(True)  # グリッド線を追加
    ax.set_xlim(*xlim)
    ax.set_ylim(*ylim)

    # X軸の箱ひげ図
    ax_box_x = fig.add_subplot(gs[2, 1], sharex=ax)
    sns.boxplot(data=df, x=ax_label_X, ax=ax_box_x)
    ax_box_x.axis('off')
    annotate_stats(ax_box_x, df[ax_label_X], 'horizontal', fontsize=10)

    # Y軸の箱ひげ図
    ax_box_y = fig.add_subplot(gs[1, 0], sharey=ax)
    sns.boxplot(data=df, y=ax_label_Y, ax=ax_box_y)
    ax_box_y.axis('off')
    annotate_stats(ax_box_y, df[ax_label_Y], 'vertical', fontsize=10)

    plt.tight_layout()
    plt.show()

# データの準備
data = {
    "X": [random.random()*1.0 for _ in range(100)],
    "Y": [random.random()*1.0 for _ in range(100)]
}
df = pd.DataFrame(data)

# 関数を呼び出してプロットを作成
plot_data(df, ax_label_X="X", ax_label_Y="Y")
```
