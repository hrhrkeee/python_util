```python
import matplotlib.pyplot as plt
import numpy as np

cmap = plt.get_cmap("tab10")

print(type(cmap), cmap.name, cmap.N)
print(cmap.colors)

for i in range(cmap.N):
    rgb_a    = np.array(cmap(i))*255
    rgb      = rgb_a[:3].astype(int)
    html_rgb = '#{:02x}{:02x}{:02x}'.format(*tuple(rgb))
    print(rgb_a,rgb,html_rgb)
```

```python
import matplotlib.pyplot as plt

cmap = plt.get_cmap("tab10")
print(cmap(0))
print(type(cmap(0)))



n = 20
for i in range(n):
    x = (i, i+1)
    y = (i, i)
    plt.plot(x, y, "o-", label=i)

plt.vlines(n/2, 0, 20, "gray", "--")
plt.legend(ncol=2)

```

```python
qualitative_colormaps = ['Pastel1', 'Pastel2', 'Paired', 
                         'Accent', 'Dark2',
                         'Set1', 'Set2', 'Set3', 
                         'tab10', 'tab20', 'tab20b','tab20c']

for cmap_num, cmap_name in enumerate(qualitative_colormaps):
    cm = plt.get_cmap(cmap_name)
    for i in range(cm.N):
        plt.barh(cmap_num, 1, color=cm(i),linewidth=5, left=i)

plt.xticks([])
plt.yticks(range(len(qualitative_colormaps)), qualitative_colormaps)

```
