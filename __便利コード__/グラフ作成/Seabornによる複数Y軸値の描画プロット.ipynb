```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import cv2
import random
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import japanize_matplotlib
from matplotlib.backends.backend_agg import FigureCanvasAgg


data_x = []
data_y = []
data_class = []
for x in np.linspace(0, 2 * np.pi, 100):
    for l in range(10):
        data_x.append(x)
        data_y.append(-1 ** 2 * 1000 *np.cos(1 * x - 2)+np.random.normal(-500, 500))
        data_class.append(random.choice([0, 1, 2, 3]))

data = {
    "x": data_x,
    "y": data_y,
    "class": data_class,
}

df = pd.DataFrame(data)
print(df)


sns.relplot(x='x', y='y', hue="class", data=df, kind='scatter', palette="tab10")
plt.show()
sns.relplot(x='x', y='y', hue="class", data=df, kind='line', palette="tab10")
plt.show()
sns.relplot(x='x', y='y', hue="class", data=df, kind='line', errorbar='sd', palette="tab10")
plt.show()
sns.relplot(x='x', y='y', hue="class", data=df, kind='line', errorbar=None, palette="tab10")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# データの読み込み
data = sns.load_dataset("penguins")

# 新しいfigureを作成
plt.figure(figsize=(12, 5))

# 1つ目のsubplotを作成
plt.subplot(1, 2, 1)  # 1行2列の1番目の位置にプロット
sns.scatterplot(x="flipper_length_mm", y="body_mass_g", data=data, hue="species")

# 2つ目のsubplotを作成
plt.subplot(1, 2, 2)  # 1行2列の2番目の位置にプロット
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="species")

# グラフの表示
plt.tight_layout()  # グラフの間隔を調整
plt.show()
```
