```python
import japanize_matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_multiaxis_boxplots(dfs:list, base_swarm_size=5):
    num_dfs = len(dfs)
    fig, axes = plt.subplots(1, num_dfs, figsize=(6*num_dfs, 6))
    
    if not isinstance(axes, np.ndarray): axes = np.array([axes])
    
    cmap = plt.get_cmap("Set3")
    
    for i, df in enumerate(dfs):
        num_cols = len(df.columns)
        
        sns.boxplot(data=df, width=0.4, ax=axes[i], color=cmap.colors[i])  # widthで箱の幅を指定
        sns.swarmplot(data=df, color=".25", ax=axes[i], size=base_swarm_size*(0.9**(num_cols+1)))

        # グリッド線を追加
        axes[i].grid(True, which='both', linestyle='--', linewidth=0.5)
        
        
        # 最大値・最小値・四分位数の数値をグラフ上に表示
        label_x_offset = -0.3*(1.1**(num_cols-1))
        for j, col in enumerate(df.columns):
            stats = df[col].describe()
            # Y座標は四分位数や最大・最小の値を使用、X座標はi（カテゴリのインデックス）を使用
            axes[i].text(j+label_x_offset, stats['min'], f'{stats["min"]:.2f}', ha='center', va='center', size=12*(0.9**(num_cols+1)))
            axes[i].text(j+label_x_offset, stats['25%'], f'{stats["25%"]:.2f}', ha='center', va='center', size=12*(0.9**(num_cols+1)))
            axes[i].text(j+label_x_offset, stats['50%'], f'{stats["50%"]:.2f}', ha='center', va='center', size=12*(0.9**(num_cols+1)))
            axes[i].text(j+label_x_offset, stats['75%'], f'{stats["75%"]:.2f}', ha='center', va='center', size=12*(0.9**(num_cols+1)))
            axes[i].text(j+label_x_offset, stats['max'], f'{stats["max"]:.2f}', ha='center', va='center', size=12*(0.9**(num_cols+1)))
        
    plt.show()



# データを作成 (ここではランダムなデータを使用)
data1 = np.random.randn(100)
data2 = np.random.randn(100)*0.2 + 20  # オフセットを追加して異なる分布を作成

df1 = pd.DataFrame({'Category 1': np.random.randn(100)})
df2 = pd.DataFrame({'Category 2': np.random.randn(100)*0.2})
df3 = pd.DataFrame({'Category 3-1': np.random.randn(100)*1.8, 'Category 3-2': np.random.randn(100)*1.8+10})
df4 = pd.DataFrame({'Category 4-1': np.random.randn(100)*1.8, 'Category 4-2': np.random.randn(100)*1.8+10, 'Category 4-3': np.random.randn(100)*1.8+10, 'Category 4-4': np.random.randn(100)*1.1+20})

# plot_multiaxis_boxplots([df1])
plot_multiaxis_boxplots([df1, df2, df3, df4])

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 仮想データを生成
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 1.5, 100)

# データの基本統計量を計算
def calculate_statistics(data):
    min_val = np.min(data)
    max_val = np.max(data)
    median = np.median(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    return min_val, max_val, median, q1, q3

stats1 = calculate_statistics(data1)
stats2 = calculate_statistics(data2)

# Seabornのboxplotを使って横並びに表示
plt.figure(figsize=(10, 6))
sns.boxplot(data=[data1, data2])
plt.title('Boxplot of Two Datasets with Statistics')

# 統計量をプロット上に表示
for i, stats in enumerate([stats1, stats2]):
    plt.text(i, stats[0], f'Min: {stats[0]:.2f}', ha='center', va='top')
    plt.text(i, stats[1], f'Max: {stats[1]:.2f}', ha='center', va='bottom')
    plt.text(i, stats[2], f'Median: {stats[2]:.2f}', ha='center', va='center')
    plt.text(i, stats[3], f'Q1: {stats[3]:.2f}', ha='center', va='top')
    plt.text(i, stats[4], f'Q3: {stats[4]:.2f}', ha='center', va='bottom')

plt.xticks([0, 1], ['Data1', 'Data2'])
plt.show()
```

```python
import japanize_matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_multiaxis_boxplots(dfs:list, base_swarm_size=5, base_figsize=(6, 10), xlim=(0, 100)):
    num_dfs = len(dfs)
    fig, axes = plt.subplots(1, num_dfs, figsize=(base_figsize[0]*num_dfs, base_figsize[1]))
    
    if not isinstance(axes, np.ndarray): axes = np.array([axes])
    
    cmap = plt.get_cmap("Set3")
    
    for i, df in enumerate(dfs):
        num_cols = len(df.columns)
        
        sns.boxplot(data=df, width=0.4, ax=axes[i], color=cmap.colors[i], orient='h')  # 水平方向に箱ひげ図を描画
        sns.swarmplot(data=df, color=".25", ax=axes[i], size=base_swarm_size*(0.9**(num_cols+1)), orient='h')  # 水平方向のswarmplot

        # グリッド線を追加
        axes[i].grid(True, which='both', linestyle='--', linewidth=0.5)
        
        # 最大値・最小値・四分位数の数値をグラフ上に表示
        label_y_offset = -0.3*(1.1**(num_cols-1))
        for j, col in enumerate(df.columns):
            stats = df[col].describe()
            # X座標は四分位数や最大・最小の値を使用、Y座標はj（カテゴリのインデックス）を使用
            axes[i].text(stats['min'], j+label_y_offset, f'{stats["min"]:.2f}', ha='center', va='center', size=12*(0.9**(num_cols+1)), rotation=90)
            axes[i].text(stats['25%'], j+label_y_offset, f'{stats["25%"]:.2f}', ha='center', va='center', size=12*(0.9**(num_cols+1)), rotation=90)
            axes[i].text(stats['50%'], j+label_y_offset, f'{stats["50%"]:.2f}', ha='center', va='center', size=12*(0.9**(num_cols+1)), rotation=90)
            axes[i].text(stats['75%'], j+label_y_offset, f'{stats["75%"]:.2f}', ha='center', va='center', size=12*(0.9**(num_cols+1)), rotation=90)
            axes[i].text(stats['max'], j+label_y_offset, f'{stats["max"]:.2f}', ha='center', va='center', size=12*(0.9**(num_cols+1)), rotation=90)
            
        if xlim is not None: axes[i].set_xlim(xlim)
        
    plt.show()



# データを作成 (ここではランダムなデータを使用)
data1 = np.random.randn(100)
data2 = np.random.randn(100)*0.2 + 20  # オフセットを追加して異なる分布を作成

df1 = pd.DataFrame({'Category 1': np.random.randn(100)})
df2 = pd.DataFrame({'Category 2': np.random.randn(100)*0.2})
df3 = pd.DataFrame({'Category 3-1': np.random.randn(100)*1.8, 'Category 3-2': np.random.randn(100)*1.8+10})
df4 = pd.DataFrame({'Category 4-1': np.random.randn(100)*1.8, 'Category 4-2': np.random.randn(100)*1.8+10, 'Category 4-3': np.random.randn(100)*1.8+10, 'Category 4-4': np.random.randn(100)*1.1+20})

# plot_multiaxis_boxplots([df1])
plot_multiaxis_boxplots([df1, df2, df3, df4])

plot_multiaxis_boxplots([df1])

```
