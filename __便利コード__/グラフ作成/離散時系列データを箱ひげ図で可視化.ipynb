```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# 12ヶ月分のデータを生成し、それぞれ30地点の気温をランダムに割り当てます。

# 各月のデータフレームを格納するリスト
monthly_data = []

for month in range(1, 120):
    # 月をラジアンに変換（1月を0とし、12月を2πとする）
    radians = (month - 1) * (2 * np.pi / 12)
    # 正弦関数に従う平均気温を計算
    # ここでは振幅を10とし、中心を20度に設定しています（気温が10から30度の範囲で変動）
    mean_temperature = 20 + 20 * np.sin(radians)
    # 平均気温の周りで標準偏差を5とする正規分布からランダムな気温データを生成
    temperatures = np.random.normal(mean_temperature, 10, 30)
    
    # 月と気温のデータフレームを作成
    df = pd.DataFrame({'Temperature': temperatures, 'Month': [month] * 30})
    monthly_data.append(df)

# 全月のデータを結合
data = pd.concat(monthly_data)
print(data)

# ボックスプロットを作成
plt.figure(figsize=(30, 6))  # グラフのサイズ指定
sns.boxplot(x='Month', y='Temperature', data=data)

# 軸ラベルの設定
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Monthly Temperature at 30 Locations')

# グラフの表示
plt.show()
```
