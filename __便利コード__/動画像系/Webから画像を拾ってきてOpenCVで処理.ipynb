```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import numpy as np
import cv2, tempfile, requests
import matplotlib.pyplot as plt

def show_img(img, dpi=150, title=None):
    fig, ax = plt.subplots(nrows=1, ncols=1, dpi=dpi)
    ax.set_title( title, fontsize=16, color='black')
    ax.axis('off')
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    return fig, ax

def imread_web(url):
    res = requests.get(url)
    image_data = np.frombuffer(res.content, dtype=np.uint8)
    img = cv2.imdecode(image_data, cv2.IMREAD_COLOR)
    return img
```

```python
# sample_img_url = "https://gahag.net/img/201607/05s/gahag-0103196580-1.jpg"
sample_img_url = "https://picsum.photos/512"
img = imread_web(sample_img_url)

show_img(img)
```
