```python
import numpy as np
import matplotlib.pyplot as plt

def generate_noise_image_rgb(width, height, mean=(127, 127, 127), std=(40, 40, 40)):
    """
    Generate an RGB noise image following a normal distribution.

    Parameters:
    - width (int): Width of the noise image.
    - height (int): Height of the noise image.
    - mean (tuple): Mean of the normal distribution for R, G, and B.
    - std (tuple): Standard deviation of the normal distribution for R, G, and B.

    Returns:
    - np.array: The RGB noise image.
    """
    noise_image_r = np.random.normal(mean[0], std[0], (height, width))
    noise_image_g = np.random.normal(mean[1], std[1], (height, width))
    noise_image_b = np.random.normal(mean[2], std[2], (height, width))

    # 0-255
    noise_image_r = np.clip(noise_image_r, 0, 255)
    noise_image_g = np.clip(noise_image_g, 0, 255)
    noise_image_b = np.clip(noise_image_b, 0, 255)

    # RGB
    noise_image = np.stack([noise_image_r, noise_image_g, noise_image_b], axis=-1).astype(np.uint8)
    return noise_image

# RGB
noise_img_rgb = generate_noise_image_rgb(256, 256)

# 
plt.imshow(noise_img_rgb)
plt.title("RGB Noise Image")
plt.axis('off')  # 
plt.show()
```
