```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import matplotlib.pyplot as plt
from pathlib import Path
import cv2

def show_img(img, dpi=150, title=None):
    fig, ax = plt.subplots(nrows=1, ncols=1, dpi=dpi)
    ax.set_title( title, fontsize=16, color='black')
    ax.axis('off')
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    return fig, ax

def resize_image(img, width=None, height=None):
    """"""
    if width is None and height is None:
        return img
    elif width is None:
        ratio = height / img.shape[0]
        width = int(img.shape[1] * ratio)
    elif height is None:
        ratio = width / img.shape[1]
        height = int(img.shape[0] * ratio)
    else:
        pass
    return cv2.resize(img, (width, height))


img_paths = list(Path("../sample_data/").glob("*.jpg"))
print(img_paths)

for img_path in img_paths:
  img = cv2.imread(str(img_path))
  r_img = resize_image(img, width=256, height=256)
  show_img(img)
  show_img(r_img)

```
