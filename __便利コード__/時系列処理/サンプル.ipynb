```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA

# yfinanceを使って為替データを取得 (USD/JPY)
ticker = "JPY=X"
start_date = "2020-01-01"
end_date = "2023-12-31"

# データ取得
forex_data = yf.download(ticker, start=start_date, end=end_date)
forex_data = forex_data[["Close"]].rename(columns={"Close": "ExchangeRate"})
forex_data.index.name = "Date"

# 時系列データをプロット
plt.figure(figsize=(12, 6))
plt.plot(forex_data.index, forex_data["ExchangeRate"], label="USD/JPY Exchange Rate")
plt.title("USD/JPY Exchange Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Exchange Rate")
plt.legend()
plt.show()

# 1. 時系列分解 (トレンドと季節性を分離する)
decomposition = seasonal_decompose(forex_data["ExchangeRate"], model="additive", period=252)
decomposition.plot()
plt.suptitle("Seasonal Decomposition", fontsize=16)
plt.show()

# 2. 定常性の検定 (データが定常かどうかを判定する)
# Augmented Dickey-Fuller (ADF) Test
adf_test = adfuller(forex_data["ExchangeRate"])
print("ADF Statistic:", adf_test[0])
print("p-value:", adf_test[1])
print("Critical Values:", adf_test[4])
if adf_test[1] < 0.05:
    print("データは定常です。")
else:
    print("データは非定常です。")

# 3. 自己相関 (ACF) と部分自己相関 (PACF) プロット
# ACF: 時系列データの自己相関を見る
plot_acf(forex_data["ExchangeRate"], lags=50)
plt.title("Autocorrelation (ACF)")
plt.show()

# PACF: 時系列データの部分自己相関を見る
plot_pacf(forex_data["ExchangeRate"], lags=50)
plt.title("Partial Autocorrelation (PACF)")
plt.show()

# 4. 自動モデル選択と予測 (ARIMA)
# auto_arima: ARIMAモデルの最適なパラメータを自動選択する
arima_model = auto_arima(forex_data["ExchangeRate"], seasonal=False, trace=True)
print("Auto-ARIMA selected order:", arima_model.order)

# ARIMA モデルのフィッティングと予測
model = ARIMA(forex_data["ExchangeRate"], order=arima_model.order)
result = model.fit()
print(result.summary())

# 予測のプロット
forecast = result.forecast(steps=30)  # 30日先を予測
plt.figure(figsize=(12, 6))
plt.plot(forex_data.index, forex_data["ExchangeRate"], label="Original")
plt.plot(pd.date_range(forex_data.index[-1], periods=30, freq="D"), forecast, label="Forecast")
plt.title("ARIMA Forecast")
plt.xlabel("Date")
plt.ylabel("Exchange Rate")
plt.legend()
plt.show()
```
