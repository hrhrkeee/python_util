```python
import datetime
import pandas as pd
import yfinance as yf
import mplfinance as mpf

data_name = 'BTC-USD'
end       = datetime.date.today()
start     = end - datetime.timedelta(weeks=5)

#データのダウンロード
df = yf.download(tickers=data_name, start=start, end=end)
df.head()

mpf.plot(df, type='candle', style='yahoo', title=data_name)
```

```python
data_name  = 'USDJPY=X'
end        = datetime.date.today()
start      = end - datetime.timedelta(days=365*20)

df = yf.download("USDJPY=X", start, end)

mpf.plot(df, type='line', title=data_name)
```
