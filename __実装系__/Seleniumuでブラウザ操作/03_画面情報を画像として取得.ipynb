```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import matplotlib.pyplot as plt
import cv2
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
import time
from pathlib import Path
from PIL import Image
import numpy as np
import io

def show_img(img, dpi=150, title=None):
    fig, ax = plt.subplots(nrows=1, ncols=1, dpi=dpi)
    ax.set_title( title, fontsize=16, color='black')
    ax.axis('off')
    ax.imshow(img)
    return fig, ax
```

```python
# Firefoxを開く
# driver = webdriver.Firefox()
# Chromeを開く
driver = webdriver.Chrome()
# Edgeを開く
# driver = webdriver.Edge()

driver.get('https://www.google.co.jp/')

# 全画面でchromeを開く
# driver.fullscreen_window()
# 最大ウィンドウでchromeを開く
# driver.maximize_window()

# スクリーンショットを撮る
now = time.strftime('%Y%m%d_%H%M%S')
output_dir = Path('./screenshot')
if not output_dir.exists():
    output_dir.mkdir()

# np.ndarrayとしてスクリーンショットを取得
screenshot = driver.get_screenshot_as_png()
data = driver.get_screenshot_as_png()
screen_np_img = np.asarray(Image.open(io.BytesIO(data))) # RGB

show_img(screen_np_img)

time.sleep(2.0)

# ブラウザを閉じる
driver.quit()
```
