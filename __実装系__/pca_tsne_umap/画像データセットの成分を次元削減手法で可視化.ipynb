### 画像のピクセルでPCA分析

```python
data_list = [
    {
        "name": "Parade",
        "path": Path("../../sample_data/WIDER_OpenData/0--Parade/"),
    },
    {
        "name": "Handshaking",
        "path": Path("../../sample_data/WIDER_OpenData/1--Handshaking/"),
    },
    {
        "name": "Demonstration",
        "path": Path("../../sample_data/WIDER_OpenData/2--Demonstration/"),
    },
    {
        "name": "Riot",
        "path": Path("../../sample_data/WIDER_OpenData/3--Riot/"),
    },
]

# t-sne, umapは計算に時間がかかるので、画像サイズを落とす
features_list = [utils.ImageDataset(data["path"], img_size=(128, 128)).get_all_tensor() for data in data_list]
plotter = utils.TensorPlotter(features_list, labels=[data["name"] for data in data_list])
plotter.multi_plot(dim=2, methods=["pca", "tsne", "umap"])
plotter.multi_plot(dim=3, methods=["pca", "tsne", "umap"])
```
