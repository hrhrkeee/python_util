```python
import numpy as np
import random
import matplotlib.pyplot as plt

import barcode
from barcode.writer import ImageWriter, ImageDraw, ImageFont

def show_img(img, dpi=150, title=None):
    fig, ax = plt.subplots(nrows=1, ncols=1, dpi=dpi)
    ax.set_title( title, fontsize=16, color='black')
    ax.axis('off')
    ax.imshow(img)
    return fig, ax

# 12桁の数字をランダムで生成
code_num = str(random.randint(450000000000, 459999999999))
bar_code = barcode.get_barcode_class('JAN') (code_num, writer=ImageWriter()) 


# 画像情報として取得
barcode_PIL_img = bar_code.render()
barcode_numpy_img = np.asarray(barcode_PIL_img) # RGB
show_img(barcode_numpy_img)

# 画像として保存
bar_code.save('output/barcode.png')
```
